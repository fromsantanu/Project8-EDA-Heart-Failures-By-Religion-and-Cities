# Heart Failures by Religion and Cities

Incidents of heart failures observed among various religions in India along with incidents happening in different cities 

### Data Preparation (Synthetic)

```python
import pandas as pd

# Creating a DataFrame for deaths due to heart failure by religion
religion_data = {
    'Religion': ['Hindu', 'Muslim', 'Christian', 'Sikh', 'Other'],
    'Deaths': [50, 20, 15, 10, 5]
}
religion_df = pd.DataFrame(religion_data)

# Saving the data to a CSV file
religion_df.to_csv('heart_failure_deaths_by_religion.csv', index=False)

# Creating a DataFrame for deaths due to heart failure by city
city_data = {
    'City': ['Mumbai', 'Delhi', 'Bangalore', 'Chennai', 'Kolkata'],
    'Deaths': [30, 25, 20, 15, 10]
}
city_df = pd.DataFrame(city_data)

# Saving the data to a CSV file
city_df.to_csv('heart_failure_deaths_by_city.csv', index=False)

print("Data files have been generated and saved.")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the data from the CSV files
religion_df = pd.read_csv('heart_failure_deaths_by_religion.csv')
city_df = pd.read_csv('heart_failure_deaths_by_city.csv')

# Create a figure with two subplots side by side
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Plotting the pie chart for deaths by religion
axs[0].pie(religion_df['Deaths'], labels=religion_df['Religion'], autopct='%1.1f%%', startangle=140)
axs[0].set_title('Deaths Due to Heart Failure by Religion in India')

# Plotting the pie chart for deaths by city
axs[1].pie(city_df['Deaths'], labels=city_df['City'], autopct='%1.1f%%', startangle=140)
axs[1].set_title('Deaths Due to Heart Failure by City in India')

# Display the plot
plt.tight_layout()
plt.show()
```
